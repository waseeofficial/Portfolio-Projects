```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

**load the csv**

```python
df = pd.read_csv('titanic_full.csv')
```

```python
df
```

**Check the null values**

```python
df.isnull().sum()
```

**Plot the data**

```python
fig, ax = plt.subplots(figsize=(4.1, 3))

ax.scatter(df.Fare, df.Survived, marker='+', color='red')

plt.show()
```

**Separate the Features a target in x and y**

```python
x = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
x
```

**Split the train and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Fare']], df.Survived, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

**Make the model Binary logistic regression model**

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[100]])
```

**Check the model score**

```python
model.score(X_test, y_test)
```
