```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset from weather.csv
df = pd.read_csv('Weather.csv')
```

```python
# Features and target variable
X = df[['Weather', 'Humidity', 'Wind']]
y = df['Play']
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to categorical features
for col in ['Weather', 'Humidity', 'Wind']:
    X[col] = le.fit_transform(X[col]) # Fit and transform each categorical column
```

```python
# Initialize Decision Tree
clf = DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X, y)
```

```python
# Plot the decision tree
plt.figure(figsize=(10, 7))
plot_tree(clf, feature_names=['Weather', 'Humidity', 'Wind'], class_names=['No', 'Yes'], filled=True)
plt.show()
```
